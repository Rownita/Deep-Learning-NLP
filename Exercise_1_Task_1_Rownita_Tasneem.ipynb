# a. Determining and Comparing the Vocabulary Sizes in these Models

I have selected English language for further analysis. The comparison of sub-word tokenization of two types of pre-trained language models: one monolingual model trained specifically on a target language and one multilingual model trained on multiple languages(including the target langugae)

### 1. English Models:
Monolingual Model: bert-base-cased

Vocabulary Size: 30000 tokens

Language Case: Trained Specially on English

Multilingual Model: bert-base-multilingual-cased

Vocabulary Size: 110,000 tokens

Languages: Trained on 104 languages, including English

The English monolingual model has a smaller vocabulary size of 30000 tokens, which reflects the fact that English is less morphologically complex compared to Finnish. The multilingual model has a significantly larger vocabulary size of 110,000 tokens. This is because it is trained on 104 languages, requiring a much larger vocabulary to represent tokens and subwords across a wide range of linguistic systens, from highly inflected languages to less inflected languages like English

The multilingual model is trained on a diverse set of 104 languages. To accomodate this wide variety, the model´s vocabulary is much larger to include subwords from different linguistic systems (e.g. scripts, word formations, inflections). Since the 110,000 tokens are shared across all 104 languages, the tokenization process for any individual language may result in more subword splits compared to monolingual models, which can dedicate their entire vocabulary to one language.

```python
from transformers import BertTokenizer # importing BertTokenizer class from the transformers library.  This class is used for tokenizing text with BERT models
```

```python
# Load the Tokenizers for the monolingual and multilingual models
# from_pretrained() method loads the tokenizer configuration and vocabulary from a pre-trained model
mono_tokenizer = BertTokenizer.from_pretrained("bert-base-cased") ## loads the tokenizer for the BERT base model with case sensitivity
multi_tokenizer = BertTokenizer.from_pretrained("bert-base-multilingual-cased") ## loads the tokenizer for the multilingual BERT model (bert-base-multilingual-cased), which can handle multiple languages
```

```python
### example text to be tokenized
text = "Finland is a Northern European nation bordering Sweden, Norway and Russia. Its capital, Helsinki, occupies a peninsula and surrounding islands in the Baltic Sea. Helsinki is home to the 18th-century sea fortress Suomenlinna, the fashionable Design District and diverse museums. The Northern Lights can be seen from the country's Arctic Lapland province, a vast wilderness with national parks and ski resorts. "
```

```python
# tokenize the text using both models
mono_tokens = mono_tokenizer.tokenize(text)
multi_tokens = multi_tokenizer.tokenize(text)
```

### mono_tokenizer.tokenize(text) tokenizes the text using the monolingual (English-specific) BERT tokenizer.
### multi_tokenizer.tokenize(text) tokenizes the same text using the multilingual BERT tokenizer.

```python
## display the tokenized output for both models
print("Monolingual Model Tokens (Bert-base-cased):")
print(mono_tokens)
```

```python
print("Multilingual Model Tokens (BERT-base-multilingual-cased):")
print(multi_tokens)
```

In the monolingual model, the entire words are retained as tokens, because this tokenizer is fine-tuned specially for English. In the multilingual model, the tokenizer splits some words into subword components (like bordering into "border", "#ing"). This occurs because the model needs to efficiently handle multiple languages with a shared vocabulary.

Tokenization is the process of breaking the text into smaller chunks called tokens. These tokens can represent words, parts of words, or punctuation marks. The provided example in the screenshots describes Finland, breaking the sentences into different colored token chunks. Different models tokenize text differently . GPT-4 and GPT-3 might break down the same text into varying tokens due to differences in how they process language.

The given screenshots are descriptions of Finland and these are tokenized by GPT models, including its borders, capital (Helsinki), some highlights like Northern lights and Lapland province.

Each token is color-coded, which helps visualize the segmentation of the sentence into smaller pieces. This can help in understanding how the model processes specific words or combination of characters. For instance, words like “Sweden” and “Russia” are counted as single tokens, while "Suomenlinna," due to its complexity, may be split into multiple tokens depending on the model.


GPT-3 and GPT-4 use a tokenization method called Byte Pair Encoding (BPE). BPE breaks down words into smaller subword units, especially when the word is not frequently seen during the training process.

The tokenizer might not have "Finland" stored as a single token in its vocabulary, so it breaks it down into smaller, more frequent subword units. For instance, it could split "Fin-" and "-land."

This subword tokenization approach allows the model to handle a wide range of languages and words, even if it has not seen a specific word in training. By breaking down the word into smaller parts (subwords), the model can still generate meaningful predictions or outputs.

Splitting words into subword tokens allows the model to be efficient with a limited vocabulary. It balances the need to capture common words in one token while still being flexible enough to process less common or compund words by breaking them into smaller parts
